## plot rewards for the language learning and language emergence scenario


```python
import pandas as pd
import seaborn as sns
```

```python
setups = ['learning', 'emergence-default-receiver', 'emergence-default-sender']

conditions = ['default', 'color', 'scale', 'shape', 'all']
vs = 4
nruns = 10
modes = ['language_learning_train_vision', 
         'language_emergence_train_vision', 
         'language_emergence_train_vision']

train_rewards = {'value':[], 'condition': [], 'setup': []}
val_rewards = {'value':[], 'condition': [], 'setup': []}

for i, setup in enumerate(setups): 
    if setup == 'learning':
        agent = 'receiver'
        load_conditions = [cond + '_default' for cond in conditions]
    elif setup == 'emergence-default-receiver': 
        agent = 'receiver'
        load_conditions = ['default'] + [conditions[i] + '_default' for i in range(1,len(conditions))]
    elif setup == 'emergence-default-sender':
        agent = 'sender'
        load_conditions = ['default'] + ['default_' + conditions[i] for i in range(1,len(conditions))]
    for c, condition in enumerate(conditions): 
        for r in range(nruns): 
            R_train = np.load('results/' + modes[i] + '/' + load_conditions[c] + str(r) + '/vs' + str(vs) + 
                     '_ml3/reward.npy')[-1]
            R_val = np.load('results/' + modes[i] + '/' + load_conditions[c] + str(r) + '/vs' + str(vs) + 
                     '_ml3/test_reward.npy')[-1]
            train_rewards['setup'].append(setup)
            val_rewards['setup'].append(setup)
            train_rewards['condition'].append(condition)
            val_rewards['condition'].append(condition)
            train_rewards['value'].append(R_train)
            val_rewards['value'].append(R_val)
            
train_reward = pd.DataFrame(data=train_rewards)
val_reward = pd.DataFrame(data=val_rewards)
```

```python
fontsize = 17
plt.figure(figsize=(18,4))

for i, rewards in enumerate([train_rewards, val_rewards]):
    plt.subplot(1,2,i+1)
    ax = sns.boxplot(x="setup", y="value", hue="condition", data=rewards)
    ax.set_xlabel('')
    ax.set_xticklabels(['learning \ndefault receiver', 'emergence \ndefault receiver', 'emergence \ndefault sender'], 
                       fontsize=fontsize-2)
    ax.set_ylabel('R', fontsize=fontsize)
    plt.yticks(fontsize=fontsize-2)
    legend = plt.legend(fontsize=fontsize-2, loc='upper right', bbox_to_anchor=(1.35,0.9), fancybox=True, 
                        shadow=True, title='other agent CNN')
    legend.get_title().set_fontsize(fontsize-2)
    plt.title(['training reward', 'test reward'][i], fontsize=fontsize)
    if i == 0:
        plt.legend().remove()
    plt.hlines(0.91, 0, 2, color='white')
plt.tight_layout()
```

```python
all_vals = []
for setup in setups: 
    for condition in conditions: 
        all_vals.append(np.mean(train_reward[(val_reward.condition==condition) & (train_reward.setup==setup)]))     
print('training', np.min(all_vals), np.max(all_vals))

all_vals = []
for setup in setups: 
    for condition in conditions: 
        all_vals.append(np.mean(val_reward[(val_reward.condition==condition) & (val_reward.setup==setup)]))
print('validation', np.min(all_vals), np.max(all_vals))
```
