# evolutionary analysis

We create the reward matrices for the evolutionary analysis and print the bootstrapped confidence intervals for each entry.

```python
from utils.reward_matrix import combined_table, show_table, bootstrapped_cis, mixed_table
```

### basic conditions

default, color, scale, shape, all

```python
conditions = ['default', 'color', 'scale', 'shape', 'all']
combinations = conditions + [(c + '_' + d) for c in conditions for d in conditions if c!=d] 

table = combined_table(combinations, n_runs=20)
mean_intervals, means, cis = bootstrapped_cis(combinations, n_runs=20)

show_table(combinations, table, upper=False)
```

```python
mean_intervals
```

### mixed trait models

all mixed models: color-scale, color-shape, scale-shape
and all relevance conditions: color irrelevant, scale irrelevant, shape irrelevant

```python
conditions = ['color-scale', 'color-shape', 'scale-shape']
combinations = conditions + [(c + '_' + d) for c in conditions for d in conditions if c!=d] 

table = combined_table(combinations, n_runs=20, mode='language_emergence_color_irrelevant')
mean_intervals, means, cis = bootstrapped_cis(combinations, n_runs=20, mode='language_emergence_color_irrelevant')

show_table(combinations, table, title='color irrelevant', vmin=0.8)
```

```python
mean_intervals
```

## control: 2 senders, 2 receivers

```python
conditions = ['default', 'scale', 'all']
combinations = [(c + '_' + d) for c in conditions for d in conditions] 

table = combined_table(combinations, n_runs=10, mode='language_emergence_2senders-2receivers', subset=True)
mean_intervals, means, cis = bootstrapped_cis(combinations, mode='language_emergence_2senders-2receivers', n_runs=10, subset=True)

show_table(combinations, table, upper=False, subset=True)
```

```python
mean_intervals
```

## control: flexible-role agents

```python
conditions = ['default', 'scale', 'all']
combinations = ['default_default', 'default_scale', 'default_all', 'scale_scale', 'scale_all', 'all_all'] 

table = mixed_table(combinations, n_runs=10, mode='language_emergence_flexible_role', subset=True, flexible_role=True)
mean_intervals, means, cis = bootstrapped_cis(combinations, mode='language_emergence_flexible_role', 
                                              n_runs=10, subset=True, flexible_role=True)

show_table(combinations, table, upper=False, subset=True)
```
